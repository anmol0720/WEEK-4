```r
#Fitting
lm.fit=lm(medv~lstat,Boston)
lm.fit
summary(lm.fit)
```

```r
#To get the variables associated with "lm.fit"
names(lm.fit)
```

```r
#To get the cooeficiant associated with "lm.fit"
coef(lm.fit)
```

```r
#To find the confidence interval
confint(lm.fit)
```

```r
#To get the confidence and prediction interval
predict(lm.fit,data.frame(lstat=c(5,10,15)),interval='confidence')
predict(lm.fit,data.frame(lstat=c(5,10,15)),interval='prediction')
```

```r
#To Plot The data
plot(lstat,medv)
#For the regression line
abline(lm.fit,col=2)
```

```r
#Some abline and plot function modifications
par(mfrow=c(2,2))
plot(lstat,medv)
abline(lm.fit,lwd=3,col='red')
plot(lstat,medv,pch=20)
plot(lstat,medv,pch='+')
plot(1:20,1:20,pch=1:20)
```

```r
par(mfrow=c(2,2))
plot(lm.fit)
```

```r
#Here the residual() function is used to calculate the residual and rstudent() function is used to calculate the Standarad Residual
par(mfrow=c(2,2))
plot(predict(lm.fit),residuals(lm.fit))
plot(predict(lm.fit),rstudent(lm.fit))
```

```r
#To get the laverage statics plot
plot(hatvalues(lm.fit))
#It tells us which observation has the higest laverage statics
which.max(hatvalues(lm.fit))
```

```r
#For MULTIPLE LINEAR REGRESSION
lm.fit=lm(medv~lstat+age,data=Boston)
summary(lm.fit)
```

```r
#for including all
lm.fit=lm(medv~.,data=Boston)
summary(lm.fit)
```

```r
#To calculate R squared
summary(lm.fit)$r.sq
```

```r
#To calculate RSE
summary(lm.fit)$sigma
```

```r
#To compute VIF
library('car')
vif(lm.fit)
```

```r
#Taking in account all datas except age
lm.fit1=lm(medv~-age,data=Boston)
summary(lm.fit1)
```

```r
#If we want to update above in the old one
lm.fit1=update(lm.fit,~.-age)
```

```r
#To include the interaction term
summary(lm(medv~lstat*age,data=Boston))
```

```r
#For non linear regression
lm.fit2=lm(medv~lstat+I(lstat^2))
summary(lm.fit2)
```

```r
#To determine that the second model is superior to first one by Null Hypothesis
lm.fit=lm(medv~lstat)
anova(lm.fit,lm.fit2)
```

```r
#Here we can se that including lstat^2 gives a little change in the graph
par(mfrow=c(2,2))
plot(lm.fit2)
```

```r
#We Can Use Log function as well
summary(lm(medv~log(rm),data=Boston))
```
